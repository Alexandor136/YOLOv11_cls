```python
import cv2

import os
from ultralytics import YOLO

model = YOLO("C:/A600/YOLOv11/runs/classify/train3/weights/best.pt")  # load a custom model

# Замените 'http://username:password@ip_address:port/video' на URL вашей IP-камеры
url = 'http://192.168.3.114/ISAPI/Streaming/channels/101/picture?snapShotImageType=JPEG'

# Открываем видеопоток
cap = cv2.VideoCapture(url)

if not cap.isOpened():
    print("Ошибка: Не удалось подключиться к камере.")
    exit()

while True:
    # Читаем кадр из потока
    ret, frame = cap.read()
    
    if not ret:
        print("Ошибка: Не удалось получить кадр.")
        break

    # Обработка кадра 
    results = model("frame")  # predict on an image

    # Отображаем обработанный кадр
    cv2.imshow('Gray Frame', results)

    # Выход из цикла при нажатии клавиши 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Освобождаем ресурсы
cap.release()
cv2.destroyAllWindows()
```
